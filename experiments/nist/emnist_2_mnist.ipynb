# NOISE $\to$ MNIST analysis

constant thermostat with: $\gamma=0.05$:

```python
path = "noise_to_mnist_unet_ConstantThermostat_gamma_0.05_max_0.0/run"
x_0, x_1, x_test = generate_mnist_samples(path,  
                                        num_timesteps=1000,
                                        time_epsilon=0.05,  
                                        device="cuda:2")
```

```python
path = "noise_to_mnist_unet_ConstantThermostat_gamma_0.05_max_0.0/run"
experiment_dir = os.path.join(results_path,"crm", path)

mnist_grid(x_1[:100], experiment_dir, num_img=100, nrow=10, figsize=(4, 4))
mnist_classifier(x_1, experiment_dir, plot_histogram=True)
mnist_noise_bridge(path,
                   x_0, 
                   num_timesteps=1000,  
                   time_epsilon=0.05,
                   num_img=6, 
                   num_timesteps_displayed=20, 
                   save_path=experiment_dir) 

get_fid(x_1, x_test)
```

```python
def get_num_groups(channels, preferred_groups=8):
    while channels % preferred_groups != 0:
        preferred_groups -= 1
    return preferred_groups
```
