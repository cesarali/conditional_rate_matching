```python
import os 
import numpy as np
import pandas as pd
import torch
import json

from conditional_rate_matching.models.generative_models.crm import (
    CRM
)

from matplotlib import pyplot as plt
from conditional_rate_matching.configs.config_files import get_experiment_dir
```

```python
experiment_dir = get_experiment_dir(experiment_name="harz_experiment",
                                    experiment_type="crm",
                                    experiment_indentifier="loss_variance_times_mnist")

crm = CRM(experiment_dir=experiment_dir,device=torch.device("cpu"))
```

```python
os.listdir(crm.experiment_files.experiment_dir)
```

```python
loss_variance_times_files = os.path.join(crm.experiment_files.experiment_dir,'metrics_loss_mean_times_best_.json')
loss_mean_times = json.load(open(loss_variance_times_files))['loss_mean_times']

loss_variance_times_files = os.path.join(crm.experiment_files.experiment_dir,'metrics_loss_variance_times_best_.json')
loss_variance_times = json.load(open(loss_variance_times_files))['loss_variance_times']
```

```python
epochs = list(map(int,loss_variance_times.keys()))
epochs.sort()
```

```python
fig, ax = plt.subplots(ncols=2,nrows=1,figsize=(12,3))
for epoch in epochs:
    if epoch % 25 == 0:
        loss_variance_per_time = loss_variance_times[str(epoch)]
        ax[1].plot(loss_variance_per_time,label=str(epoch),alpha=0.3)
        ax[1].legend(loc="best")
    
        loss_mean_per_time =   loss_mean_times[str(epoch)]
        ax[0].plot(loss_mean_per_time,label=str(epoch),alpha=0.3)
        ax[0].legend(loc="best")
        
        loss_mean_times[str(epoch)]
```
